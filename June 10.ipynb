```python
import os 
import glob
filename = os.path.join('data','03_Prod.mdout')
f = open(filename, 'r')
data = f.readlines()
f.close()
f_write = open('totE.txt', 'w+')
for line in data:
    if 'Etot' in line:
        f_write.write(f'{line.split()[2]}\n')
f_write.close()
```
